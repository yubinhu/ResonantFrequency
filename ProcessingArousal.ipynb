```python
def sample(df):
    # assume column[0] is the time frame, so always sample it
    columns = df.columns
    l = []
    for i in range(len(columns)):
        if i%8 != 0:
            l.append(columns[i])
    df.drop(columns = l, inplace=True)
sample(feature)
print(feature.columns)
```

## Chop
Only need 15s-30s, and 30s-45

```python
feature.drop(index=feature.index[:30], axis=0, inplace=True)
feature
```

```python
def getVectors(feature):
    X1 = np.array([])
    for i in range(30):
        temp = np.array(feature.values.tolist()[i][1:])
        X1 = np.concatenate((X1, temp), axis=0)


    X2 = np.array([])
    for i in range(30,60):
        temp = np.array(feature.values.tolist()[i][1:])
        X2 = np.concatenate((X2, temp), axis=0)

    Xf = np.vstack((X1,X2))
    return Xf
print(getVectors(feature))
```

```python
# Now functionalize this whole process

def read(si):
    filename = str(si)+".csv"
    feature = pandas.read_csv(features_path+filename, header=0, sep=';')
    sample(feature)
    feature.drop(index=feature.index[:30], axis=0, inplace=True)
    return getVectors(feature)

read(3)
```

## Now handles the annotations

```python
valence = pandas.read_csv(annotation_path+"arousal_cont_average.csv", header=0, sep=',')
```

```python
valence
```

```python
GrandTotal = 300 # total number of songs we will consider
id_list = [int(valence.iloc[i]['song_id']) for i in range(GrandTotal)]
print(id_list)
```

```python
y = np.array([0])
for i in id_list:
    valencepure = valence.loc[valence['song_id'] == i].values.tolist()[0][2:]
    firsthalf = valencepure[:30]
    valence1 = sum(firsthalf)/len(firsthalf)
    secondhalf = valencepure[30:]
    valence2 = sum(secondhalf)/len(secondhalf)
    valences = np.array([[valence1],[valence2]])
    y = np.vstack((y,valences))
y = np.delete(y, 0, axis=0)
print(y)

# !!!!!!!!!!! y is finished here
```

```python
y.shape
```

## Now load features of all relevant music

```python
x = read(id_list[0])
for i in id_list[1:]:
    x = np.vstack((x, read(i)))
print(x)
```

```python
x.shape
```

# Machine Learning begins here

```python
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=23)
```

```python
X_train
```

```python
reg = Ridge()
reg.fit(X_train,y_train)

print("Ridge Regression")
print("Training Score",reg.score(X_train, y_train))
print("Testing Score",reg.score(X_test, y_test))
np.savetxt("coef.csv", reg.coef_, delimiter=",")
```

```python
from sklearn.ensemble import AdaBoostRegressor, GradientBoostingRegressor
```

```python
ada = AdaBoostRegressor()
ada.fit(X_train, y_train.ravel())
```

```python
print("Ada Boost")
print("Training Score",ada.score(X_train, y_train.ravel()))
print("Testing Score",ada.score(X_test, y_test.ravel()))
```

## I want to test the classification accuracy

```python
print(y_train)
```

```python
def accuracy(model, X, y):
    yp = model.predict(X) # model prediction
    correct_num = 0
    n = len(y)
    if len(yp)>1:
        yp = yp.reshape(len(yp),1)
    for i in range(n):
        if yp[i][0] * y[i][0] > 0:
            correct_num +=1
    return correct_num/n
    
```

```python
accuracy(ada,X_test, y_test)
```

```python
accuracy(reg,X_test, y_test)
```
